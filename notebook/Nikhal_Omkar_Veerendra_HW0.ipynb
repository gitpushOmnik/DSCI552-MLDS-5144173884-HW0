### (a) Create two one dimensional arrays x and y and plot y vs x, add title, xlabel, ylabel, grid.

### (b) Create multiple arrays and plot them with different styles, add legends, add text/mathematical equations on the plot.

We plot 3 graphs below wth equations: y=45x+1, y=3x^2, y=-60x. We also place our legend on the upper center position of the graph. We also assign linestyle and allocate colour to each graphs.

```python
x_array = np.array([50, 60, 70, 80, 90])
y_array1 = np.array([45*value + 1 for value in x])
y_array2 = np.array([3*value*value for value in x])
y_array3 = np.array([-60*value for value in x])

plt.plot(x, y_array1, label='y = 45x+1', linestyle='solid', color='r')
plt.plot(x, y_array2, label='y = 3x^2', linestyle='dashed', color='g')
plt.plot(x, y_array3, label='y = -60x', linestyle='dashdot', color='b')

plt.legend(['y=45x+1', 'y=3x^2', 'y=-60x'], loc='upper center')

plt.show()
```

### (c) Create multiple subplots, play around with the figure size, text font/size

We plot 3 subplots with the same equations as above. 

```python
x_array = np.array([50, 60, 70, 80, 90])
y_array1 = np.array([45 * value + 1 for value in x_array])
y_array2 = np.array([3 * value * value for value in x_array])
y_array3 = np.array([-60 * value for value in x_array])

plt.figure(figsize=(30, 6))

plt.subplot(161)
plt.plot(x_array, y_array1, color='r', linestyle='solid')
plt.legend(['Line y=45x+1'], loc=4)
plt.text(60, 1800, 'Line y=45x+1', fontsize=15, color='r')

plt.subplot(162)
plt.plot(x_array, y_array2, color='g', linestyle='dashed')
plt.legend(['Parabola y=3x^2'], loc=4)
plt.text(55, 3600, 'Parabola y=3x^2', fontsize=15, color='g')

plt.subplot(163)
plt.plot(x_array, y_array3, color='b', linestyle='dashdot')
plt.legend(['3rd Degree y=-60x'], loc=4)
plt.text(50, -6000, '3rd Degree y=-60x', fontsize=15, color='b')

plt.show()
```
